```python
import numpy as np
```

```python
import librosa
import librosa.display
import matplotlib.pyplot as plt
```

```python
file = 'r1.wav'

sig , sr = librosa.load(file,mono=True,sr=None)
time=np.arange(0,sig.shape[0])/sr
plt.plot(time,sig)
```

```python
sig.shape
max(abs(sig))
excursion=max(abs(sig))
sigNew = sig / excursion
max(abs(sigNew))
```

```python
plt.plot(time,sigNew)
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.title('r1 Waveform')
```

```python
plt.plot(time,sigNew)
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.title('r0 Waveform')
plt.xlim(0,1)
plt.ylim(-.18,.18)
```

```python
plt.plot(time[1:44100],sigNew[1:44100])
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.title('r0 Waveform')
```

```python
r1File = 'r1.wav'
r2File = 'r2.wav'

r1Sig , sr2 = librosa.load(r1File,mono=True,sr=None)
r2Sig , sr3 = librosa.load(r2File,mono=True,sr=None)

timer1=np.arange(0,r1Sig.shape[0])/sr2
timer2=np.arange(0,r2Sig.shape[0])/sr3
```

```python
plt.plot(timer1,r1Sig)
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.title('R1 Waveform')
```

```python
plt.plot(timer1,r1Sig)
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.title('R1 Waveform zoomed')
plt.xlim(15,25)
plt.ylim(-.5,.5)
```

```python
r1Spec = librosa.amplitude_to_db(np.abs(librosa.stft(r1Sig)), ref=np.max)
librosa.display.specshow(r1Spec, y_axis='linear')
plt.colorbar(format='%+2.0f dB')
plt.title('R1: Linear-frequency power spectrogram')
```

```python
librosa.display.specshow(r1Spec, y_axis='log')
plt.colorbar(format='%+2.0f dB')
plt.title('R1: Log-frequency power spectrogram')
```

```python
plt.plot(timer2,r2Sig)
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.title('R2 Waveform')
```

```python
plt.plot(timer2,r2Sig)
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.title('R2 Waveform zoomed')
plt.xlim(0,10)
plt.ylim(-.8,.8)
```

```python
r2Spec = librosa.amplitude_to_db(np.abs(librosa.stft(r1Sig)), ref=np.max)
librosa.display.specshow(r2Spec, y_axis='linear')
plt.colorbar(format='%+2.0f dB')
plt.title('R2: Linear-frequency power spectrogram')
```

```python
librosa.display.specshow(r2Spec, y_axis='log')
plt.colorbar(format='%+2.0f dB')
plt.title('R2: Log-frequency power spectrogram')
```
